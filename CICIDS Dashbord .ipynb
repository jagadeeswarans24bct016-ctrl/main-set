```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df["MI_dir_L0.1_mean"],df['HH_L0.1_mean'],color="red")
plt.title("MI_dir_L0.1_mean vs HH_L0.1_mean")
plt.xlabel("MI_dir_L0.1_mean")
plt.ylabel("HH_L0.1_mean")
plt.show()
```

```python
top_20_devices = df['Device_Name'].value_counts().nlargest(20)

plt.figure(figsize=(12, 6))
plt.bar(top_20_devices.index.astype(str), top_20_devices.values, color="red")
plt.title("Top 20 Most Used Destination device")
plt.xlabel("Device Name")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.hist(df['MI_dir_L0.1_mean'], bins=50)
plt.title("Histogram of MI_dir_L0.1_mean ")
plt.xlabel("MI_dir_L0.1_mean")
plt.ylabel("Frequency")
plt.show()
```

```python
counts = df['label'].value_counts()

plt.figure(figsize=(10,10))
plt.pie(counts, labels=counts.index, autopct="%1.1f%%", pctdistance=0.80)

circle = plt.Circle((0,0), 0.70, color='white')
plt.gca().add_artist(circle)

plt.legend(
    [f"{counts.index[i]} = {counts.iloc[i]}" for i in range(len(counts))],
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    fontsize=9
)

plt.title("Label Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,10))

plt.subplot(2,2,1)
plt.scatter(df['MI_dir_L0.1_mean'], df['HH_L0.1_mean'], color="red")
plt.title("MI_dir_L0.1_mean vs 'HH_L0.1_mean")
plt.xlabel("MI_dir_L0.1_mean")
plt.ylabel("'HH_L0.1_mean")

plt.subplot(2,2,2)
plt.hist(df['MI_dir_L0.1_mean'], bins=50, color="blue")
plt.title("Histogram of MI_dir_L0.1_mean")
plt.xlabel("MI_dir_L0.1_mean")
plt.ylabel("Frequency")

plt.subplot(2,2,3)
top_20_ports = df['Device_Name'].value_counts().nlargest(20)
plt.bar(top_20_ports.index.astype(str), top_20_ports.values, color="green")
plt.title("Top 20 Most Used Device Name")
plt.xlabel("Device Name")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')

plt.subplot(2,2,4)
counts = df['label'].value_counts()

wedges, texts, autotexts = plt.pie(
    counts,
    labels=None,
    autopct="%1.1f%%",
    pctdistance=0.80
)


circle = plt.Circle((0,0), 0.70, color='white')
plt.gca().add_artist(circle)

plt.title("Label Distribution")


plt.legend(
    counts.index,
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    fontsize=9
)

plt.tight_layout()
plt.savefig("Report5.png", dpi=300)
plt.show()
```
